```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_excel("DailyActivities.xlsx")
data=pd.DataFrame(data=df.values,index=df["Area of Interest"]).T
data=data.drop(index=0)
data.reset_index(drop=True,inplace=True)
data.columns.name =""
data.insert(0,"People",["Charles","Henry","Susan"],True)
small_data=data[["People","Work","Family","Homeworks","Individual","Socializing","Spare Time","Sleep"]]
small_data=small_data.melt(id_vars=["People"])
small_data.columns=["People","Area of Interest","Values"]
sns.barplot(x="People", y="Values",hue="Area of Interest", data=small_data);
```
